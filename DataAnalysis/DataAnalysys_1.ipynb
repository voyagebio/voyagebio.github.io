```python
#隐藏若干列元素，方便观察
df = pd.read_csv('C:/Users/22454/Desktop/code/data/train.csv')
df.drop(['PassengerId','Name','Age','Ticket'],axis=1).head(3)
```

```python
#以"Age"为筛选条件，显示年龄在10岁以下的乘客信息
df[df["Age"]<10].head(3)
#以"Age"为条件，将年龄在10岁以上和50岁以下的乘客信息显示出来，并将这个数据命名为midage
midage = df[(df["Age"]>10)& (df["Age"]<50)]
midage.head(3)
```

```python
#将midage的数据中第100行的"Pclass"和"Sex"的数据显示出来
midage = midage.reset_index(drop=True)
midage.head(3)
#reset_index的作用：用索引重置生成一个新的DataFrame或Series，使用drop参数来避免将旧索引添加为列
```
